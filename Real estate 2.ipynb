```python
price = []

for i in soup.find_all('div', class_="result__details__price price_click_origin"):
    price_text = i.text.strip().replace('None','')
    price.append(price_text)
for price in price:
    print(price)
```

```python
address = []

for i in soup.find_all('div', class_="listing__title__text"):
    address_text = i.text.strip().replace('None','')
    address.append(address_text)
for addr in address:
    print(addr)
```

```python
b_h_k = []

for i in soup.find_all('span', class_="rooms"):
    b_h_k_text = i.text.strip().replace('\xa0','')
    b_h_k.append(b_h_k_text)

for bhk in b_h_k:
    print(bhk)
```
