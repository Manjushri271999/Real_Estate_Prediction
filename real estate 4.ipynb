```python
import pandas as pd 
import requests
from bs4 import BeautifulSoup
```

```python
page = requests.get('https://property.trovit.co.in/zone-10-kodambakkam')
```

```python
page
```

```python
soup=BeautifulSoup(page.content)
soup
```

```python
bedrooms= soup.find('div', class_="item-property item-rooms")
bedrooms
```

```python
bedrooms.text.replace('\xa0','')
```

```python
bedrooms= []

for i in soup.find_all('div', class_="item-property item-rooms"):
     bedrooms.append(i.text.replace('\xa0', ' '))

bedrooms
```

```python
import re

# List of strings representing the number of bedrooms
bedrooms_list = [
    '   3  bedrooms ',
    '   3  bedrooms ',
    '   3  bedrooms ',
    '   2  bedrooms ',
    '   2  bedrooms ',
    '   2  bedrooms ',
    '   2  bedrooms ',
    '   2  bedrooms ',
    '   3  bedrooms ',
    '   2  bedrooms ',
    '   2  bedrooms ',
    '   3  bedrooms ',
    '   4  bedrooms ',
    '   1  bedrooms ',
    '   2  bedrooms ',
    '   4  bedrooms ',
    '   1  bedrooms ',
    '   3  bedrooms ',
    '   4  bedrooms ',
    '   3  bedrooms ',
    '   2  bedrooms ',
    '   2  bedrooms ',
    '   2  bedrooms ',
    '   2  bedrooms ',
    '   2  bedrooms '
]
bed_regex = r'\b\d+\b'

bedrooms = [int(re.search(bed_regex, s).group()) for s in bedrooms_list]

print("Number of bedrooms:", bedrooms)
print("Length of bedrooms list:", len(bedrooms))
```
