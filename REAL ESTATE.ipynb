```python
data = [
    '4 BD',
    '4 BA',
    '3 BD',
    '2 BA',
    '2 BD',
    '2 BA',
    '3 BD',
    '2 BA',
    '1 BD',
    '1 BA',
    '3 BD',
    '1 BA',
    '2 BD',
    '1.5 BA',
    '3 BD',
    '2.5 BA',
    '2 BD',
    '2 BA',
    '3 BD',
    '1 BA'
]

bedrooms = []
bathrooms = []

for item in data:
    if 'BD' in item:
        bedrooms.append(item)
    elif 'BA' in item:
        bathrooms.append(item)

print("Bedrooms:", bedrooms)
print("Bathrooms:", bathrooms)
```
